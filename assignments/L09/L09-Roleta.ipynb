##A) Vamos apostar?

Em uma roleta podemos jogar de diferentes formas, conforme vemos na figura abaixo. Assumindo uma roleta justa (sem zeros) e usando seus conhecimentos de probabilidade estime a probabilidade (de forma teórica e experimental) dos dez diferentes tipos de aposta.

**Tipos de Aposta**


1. *Color Bet*: aposta em uma das cores (vermelho/preto)
2. *Straight Up Bet*: aposta em um número específico
3. *Corner Bet*: aposta em 4 números vizinhos pelo canto do quadro em que estão
4. *Split bet*: aposta em dois números ligados pelo lado do quadro em que estão
5. *Column bet*: aposta em 12 números de uma coluna (no desenho é uma linha)
6. *Dozen bet*: aposta em 12 números em sequência (1º: 1-12, 2º: 13-24, 3º: 25-36)
7. *Even/Odd bet*: aposta nos números pares ou ímpares
8. *High/Low bet*: aposta em 18 números (1-18: baixo, 19-36: alto)
9. *Street bet*: aposta em 3 números em linha (no desenho é coluna)
10. *Line bet*: aposta em 6 números de duas linhas diferentes

### Probabilidade Experimental

Procure entender o código abaixo e modifique a função ```playRoulette``` para estimar a probabilidade de outros tipos de apostas. Se achar importante, você pode modificar o número de giros.

````python
import random

class RoletaJusta():
    def __init__(self):
        self.espacos = []
        for i in range(1,37):
            self.espacos.append(i)
        self.bola = None
    def roda(self):
        self.bola = random.choice(self.espacos)
    def ePreto(self):
        if type(self.bola) != int:
            return False
        if ((self.bola > 0 and self.bola <= 10)\
            or (self.bola>18 and self.bola<=28)):
            return self.bola%2 == 0
        else:
            return self.bola%2 == 1
    def eoNumero(self, numero):
        return self.bola == numero
    def __str__(self):
        return 'Roleta Justa'

def giraRoleta(jogo, numGiros, imprimir = True):
    NumeroApostado = 2
    totPreto, totNumeroApostado = 0.0, 0.0
    for i in range(numGiros):
        jogo.roda()
        totPreto += 1 if jogo.ePreto() else 0
        totNumeroApostado += 1 if jogo.eoNumero(NumeroApostado) else 0
    if imprimir:
        print(numGiros, 'giros do jogo', jogo)
        print('Probabilidade de aposta no preto =', str(totPreto/numGiros))
        print('Probabilidade de aposta no número =', str(totNumeroApostado/numGiros))

numGiros = 100
jogo = RoletaJusta()
giraRoleta(jogo, numGiros)
````
